```python
import pandas as pd
import numpy as np
import torch
import torch.nn.functional as F
import torch_directml

# device = torch.device('cpu')
device = torch_directml.device()

# Load the data from the URL.
df_train = pd.read_csv("./data/Training set.csv")
df_test = pd.read_csv("./data/Test set.csv")
```

```python
df_train.loc[df_train['Sex'] == 'Male',  'Sex'] = 1
df_train.loc[df_train['Sex'] == 'Female','Sex'] = 0
df_test.loc[df_test['Sex']   == 'Male',  'Sex'] = 1
df_test.loc[df_test['Sex']   == 'Female','Sex'] = 0
i_to_sex = {1: 'Male', 0: 'Female'}
sex_to_i = {'Male': 1, 'Female': 0}
```

### Remove Outliers

A common way to remove outliers is to use **z-scores**. Outliers are values that have z-scores that are greater than 3 or less than -3.

```python
z_scores = (df_train['Height'] - df_train['Height'].mean()) / df_train['Height'].std()
df_train = df_train.drop(df_train[abs(z_scores) > 3].index)
```

```python
z_scores = (df_train['Weight'] - df_train['Weight'].mean()) / df_train['Weight'].std()
df_train = df_train.drop(df_train[abs(z_scores) > 3].index)
```

```python
import matplotlib.pyplot as plt

# Define data and labels
height_data = df_train['Height'].dropna().values
weight_data = df_train['Weight'].dropna().values

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Create boxplots for height and weight
ax1.boxplot(height_data)
ax1.set_title('Height')
ax1.set_ylabel('(cm)')

ax2.boxplot(weight_data)
ax2.set_title('Weight')
ax2.set_ylabel('(kg)')

# Display the plot
plt.show()
```

```python
class NeuralNetwork(torch.nn.Module):

    def __init__(self, n_input, n_hidden, n_output):
        super().__init__()

        self.linear1 = torch.nn.Linear(n_input, n_hidden)
        self.relu = torch.nn.ReLU()
        self.linear2 = torch.nn.Linear(n_hidden, n_output)
        
        # not sure why it is not mitigating the hockey stick effect
        torch.nn.init.kaiming_uniform_(self.linear1.weight, nonlinearity='relu')
        torch.nn.init.kaiming_normal_(self.linear2.weight, nonlinearity='sigmoid')

    def forward(self, x):
        x = self.linear1(x)
        x = self.relu(x)
        x = self.linear2(x)
        # We use log_softmax since it is numerically more stable than
        # first taking the softmax and then the log
        x = F.log_softmax(x, dim=1)
        return x

# Create the model
model = NeuralNetwork(2, 10, 2).to(device)
```

```python
# Define the input and output data
train_inputs = torch.tensor(df_train[["Height", "Weight"]].values, dtype=torch.float32)
train_inputs_norm = (train_inputs - train_inputs.mean(axis=0)) / train_inputs.std(axis=0)

train_outputs = torch.tensor(df_train["Sex"].astype(int).values)
```

```python
train_inputs_norm
```

```python
print('input shape:', train_inputs.shape)
print('output shape:', train_outputs.shape)
```

```python

# Load into device (GPU)
train_inputs_norm = train_inputs_norm.to(device)
train_outputs = train_outputs.to(device)
```

```python
train_losses = []
epochs = []
```

```python
# Define the loss function
loss_fn = torch.nn.NLLLoss()

# Define the optimizer
lr = 0.001
optimizer = torch.optim.Adam(model.parameters(), lr=lr)

num_epochs = 1000

# Train the model
model.train() # see: https://cs230.stanford.edu/blog/pytorch/#training-vs-evaluation
for epoch in range(num_epochs):

    # Forward pass: compute predicted y by passing x to the model.
    output = model(train_inputs_norm)

    # Calculate the loss
    loss = loss_fn(output, train_outputs)

    # Before the backward pass, use the optimizer object to zero all of the
    # gradients for the Tensors it will update (which are the learnable weights
    # of the model)
    optimizer.zero_grad()

    # Backward pass: compute gradient of the loss with respect to model parameters
    loss.backward()

    # Calling the step function on an Optimizer makes an update to its parameters
    optimizer.step()
    
    # Record loss and epoch values
    if (epoch + 1) % 100 == 0:
        train_losses.append(loss.item())
        epochs.append(epoch)

plt.plot(np.arange(len(epochs)), train_losses)
plt.title('Training Loss over Time')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
print(train_losses[-1])
```

```python
# Evaluate

# Define the input and output data
test_inputs = torch.tensor(df_test[["Height", "Weight"]].values, dtype=torch.float32)
test_inputs_norm = (test_inputs - test_inputs.mean(axis=0)) / test_inputs.std(axis=0)

test_outputs = torch.tensor(df_test["Sex"].astype(int).values)
```

```python
# Make predictions on the test set using the model
# by converting both the predicted and true class labels to NumPy arrays explicitly,
# we can ensure that their data types and shapes match, we avoid unexpected behavior.
model.eval() # see: https://cs230.stanford.edu/blog/pytorch/#training-vs-evaluation
with torch.no_grad():
    pred = model(test_inputs_norm.to(device))
    pred = np.argmax(pred.cpu().numpy(), axis=1)

# Compute the accuracy of the predictions
test_outputs_np = test_outputs.numpy()
accuracy = np.sum(pred == test_outputs_np) / len(test_outputs_np)

# Print the accuracy
print("The accuracy of the testing set: {:.2f}%".format(accuracy * 100))
```
